# DIRECT SQUARE LATTICE

```python
import numpy as np
import matplotlib.pyplot as plt

# Primitive vector magnitudes
a = np.array([1,0])
b = np.array([0,1])  # <-- Change this to 1 if you want a true square lattice


area=np.cross(a,b)
print(area)
# Lattice size
n1 = 5
n2 = 5

# Arrays for storing coordinates
x = []
y = []
x1=np.linspace(0,4)



# Generate lattice points
for i in range(n1):
    for j in range(n2):
        r=i*a+j*b
        x.append(r[0])
        y.append(r[1])
# Plotting
plt.figure(figsize=(6,6))
plt.scatter(x, y, s=100, color='blue')  # s = size of points

plt.axhline(y=2, color='red', linestyle='-', linewidth=2)
plt.axhline(y=1, color='red', linestyle='-', linewidth=2)
plt.axhline(y=0, color='red', linestyle='-', linewidth=2)
plt.axhline(y=3, color='red', linestyle='-', linewidth=2)
plt.axhline(y=4, color='red', linestyle='-', linewidth=2)


plt.axvline(x=2, color='red', linestyle='-', linewidth=2)
plt.axvline(x=1, color='red', linestyle='-', linewidth=2)
plt.axvline(x=0, color='red', linestyle='-', linewidth=2)
plt.axvline(x=3, color='red', linestyle='-', linewidth=2)
plt.axvline(x=4, color='red', linestyle='-', linewidth=2)

plt.xticks([0,1,2,3,4])
plt.yticks([0,1,2,3,4])
plt.title('2D Square Lattice', fontsize=14)
plt.xlabel(r"$k_x$", fontsize=14)
plt.ylabel(r"$k_y$", fontsize=14)

plt.grid(True)
plt.gca().set_aspect('equal')  # Equal scaling


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Plot setup
fig, ax = plt.subplots(figsize=(6, 6))

# Set x and y limits to zoom into the Brillouin zone
ax.set_xlim(-np.pi*2, np.pi*2)
ax.set_ylim(-np.pi*2, np.pi*2)

# Draw the Brillouin zone square manually (vertices)
k = np.pi
square = np.array([
    [-k, -k],
    [ k, -k],
    [ k,  k],
    [-k,  k],
    [-k, -k]  # close the loop
])
ax.plot(square[:,0], square[:,1], color='red', linewidth=2, label='1st Brillouin Zone')

# Add reciprocal lattice points
a = 2 * np.pi * np.array([1, 0])
b = 2 * np.pi * np.array([0, 1])
x = []
y = []
for i in range(-2, 3):
    for j in range(-2, 3):
        r = i * a + j * b
        x.append(r[0])
        y.append(r[1])
ax.scatter(x, y, color='blue', s=50)

# Dashed lines at  and 2
for val in [-2*np.pi, -np.pi, 0, np.pi, 2*np.pi]:
    ax.axhline(y=val, color='gray', linestyle='--', linewidth=0.5)
    ax.axvline(x=val, color='gray', linestyle='--', linewidth=0.5)

#  ticks
tick_positions = [-2*np.pi, -np.pi, 0, np.pi, 2*np.pi]
tick_labels = [r"$-2\pi$", r"$-\pi$", r"$0$", r"$\pi$", r"$2\pi$"]
ax.set_xticks(tick_positions)
ax.set_yticks(tick_positions)
ax.set_xticklabels(tick_labels, fontsize=12)
ax.set_yticklabels(tick_labels, fontsize=12)

# Label and style
ax.set_xlabel(r"$k_x$", fontsize=14)
ax.set_ylabel(r"$k_y$", fontsize=14)
ax.set_title("First Brillouin Zone (Manual Drawing)", fontsize=14)
ax.set_aspect('equal')
ax.grid(True, linestyle='--', alpha=0.3)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-2*np.pi, 2*np.pi)
ax.set_ylim(-2*np.pi, 2*np.pi)

# Reciprocal lattice
a = 2 * np.pi * np.array([1, 0])
b = 2 * np.pi * np.array([0, 1])
x, y = [], []
for i in range(-2, 3):
    for j in range(-2, 3):
        r = i * a + j * b
        x.append(r[0])
        y.append(r[1])
ax.scatter(x, y, color='blue', s=50, label='Reciprocal Lattice Points')

# Constants
pi = np.pi
L = 2 * pi

# 1st Brillouin Zone (center cyan square)
square_patch = patches.Rectangle(
    (-pi, -pi),
    2*pi,
    2*pi,
    linewidth=2,
    edgecolor='red',
    facecolor='cyan',
    alpha=1,
    label='1st Brillouin Zone'
)
ax.add_patch(square_patch)

# Fill 4 red triangles (Second BZ regions)
# Top triangle
plt.fill([ -pi, 0,  pi], [ pi, L,  pi], color='red', alpha=1)
# Bottom triangle
plt.fill([ -pi, 0,  pi], [-pi, -L, -pi], color='red', alpha=1)
# Left triangle
plt.fill([-pi, -L, -pi], [-pi, 0,  pi], color='red', alpha=1)
# Right triangle
plt.fill([ pi,  L,  pi], [-pi, 0,  pi], color='red', alpha=1)

# Dashed  lines
for val in [-2*pi, -pi, 0, pi, 2*pi]:
    ax.axhline(y=val, color='gray', linestyle='--', linewidth=0.5)
    ax.axvline(x=val, color='gray', linestyle='--', linewidth=0.5)

#  ticks
tick_positions = [-2*pi, -pi, 0, pi, 2*pi]
tick_labels = [r"$-2\pi$", r"$-\pi$", r"$0$", r"$\pi$", r"$2\pi$"]
ax.set_xticks(tick_positions)
ax.set_yticks(tick_positions)
ax.set_xticklabels(tick_labels, fontsize=12)
ax.set_yticklabels(tick_labels, fontsize=12)

# Label high-symmetry points
ax.text(0, 0, r"$\Gamma$", fontsize=14, ha='center', va='center', color='black', weight='bold')

# X points
ax.text(np.pi, 0, r"$X$", fontsize=14, ha='left', va='center', color='black')
ax.text(-np.pi, 0, r"$X$", fontsize=14, ha='right', va='center', color='black')
ax.text(0, np.pi, r"$X$", fontsize=14, ha='center', va='bottom', color='black')
ax.text(0, -np.pi, r"$X$", fontsize=14, ha='center', va='top', color='black')

# M points
ax.text(np.pi, np.pi, r"$M$", fontsize=14, ha='left', va='bottom', color='black')
ax.text(-np.pi, np.pi, r"$M$", fontsize=14, ha='right', va='bottom', color='black')
ax.text(-np.pi, -np.pi, r"$M$", fontsize=14, ha='right', va='top', color='black')
ax.text(np.pi, -np.pi, r"$M$", fontsize=14, ha='left', va='top', color='black')

# Labels and formatting
ax.set_xlabel(r"$k_x$", fontsize=14)
ax.set_ylabel(r"$k_y$", fontsize=14)
ax.set_title("Filled First and Second Brillouin Zones", fontsize=14)
ax.set_aspect('equal')
ax.grid(True, linestyle='--', alpha=0.3)
ax.legend()
plt.tight_layout()
plt.show()
```

# BRILLOUIN ZONE FOR HEXAGONAL LATTICE


```python
import numpy as np
import matplotlib.pyplot as plt

# Define radius of 1st BZ (distance from center to vertex)
R = 4 * np.pi / 3

# Angles for 6 vertices (every 60 degrees)
angles = np.linspace(0, 2 * np.pi, 7)  # Include 0 and 2 to close the loop
x_hex = R * np.cos(angles)
y_hex = R * np.sin(angles)

# Plot hexagon
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(x_hex, y_hex, 'r-', linewidth=2, label='1st Brillouin Zone')  # Red boundary
ax.fill(x_hex, y_hex, color='cyan', alpha=0.5)  # Fill with color

# Origin
ax.plot(0, 0, 'ko', label=' (Gamma Point)')

# Symmetry points
ax.text(0.2, 0.2, r"$\Gamma$", fontsize=14, ha='center', va='center')
ax.text(R, 0, r"$M$", fontsize=14, ha='left', va='bottom')
ax.text(R/2, np.sqrt(3)*R/2, r"$K$", fontsize=14, ha='left', va='bottom')

# Formatting
ax.set_aspect('equal')
ax.set_xlim(-R * 1.2, R * 1.2)
ax.set_ylim(-R * 1.2, R * 1.2)
ax.set_xlabel(r"$k_x$")
ax.set_ylabel(r"$k_y$")
ax.set_title("First Brillouin Zone of 2D Hexagonal Lattice")
ax.grid(True, linestyle='--', alpha=0)
ax.legend()
plt.tight_layout()
plt.show()
```

# ENERGY VOLUME CURVE FOR TITANIUM

```python
import numpy as np
import matplotlib.pyplot as plt

# Given arrays
E = np.array([
    -0.15512927E+02, -0.15380455E+02, -0.15123191E+02, -0.14722995E+02,
    -0.14158092E+02, -0.13419261E+02, -0.12490846E+02,
    -0.15540481E+02, -0.15480748E+02, -0.15384324E+02, -0.15253292E+02,
    -0.14611225E+02, -0.13759103E+02, -0.12798886E+02, -0.11798886E+02
])

a = np.array([
    0.98, 0.96, 0.94, 0.92, 0.90, 0.88, 0.86,
    1.00, 1.02, 1.035, 1.05, 1.10, 1.15, 1.20, 1.25
])

# Sort arrays
idx = np.argsort(a)
a_s = a[idx]
E_s = E[idx]

# Find minimum energy point
min_i = np.argmin(E_s)
a_min, E_min = a_s[min_i], E_s[min_i]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(a_s, E_s, lw=2, color='teal', label='Energy vs $a$')
plt.scatter(a_s, E_s, s=60, facecolor='gold', edgecolor='black', zorder=3)
plt.scatter(a_min, E_min, s=100, color='crimson', edgecolor='black', label='Minimum Energy', zorder=4)

# Annotate minimum energy value near the arrow
plt.annotate(
    rf'Minimum Energy: $E = {E_min:.2f}\ \mathrm{{eV}}$',
    xy=(a_min, E_min),
    xytext=(a_min + 0.015, E_min + 2),  # Short clean arrow
    arrowprops=dict(arrowstyle='->', color='crimson', linewidth=1.5),
    fontsize=12,
    color='black'
)

# Labels & Title with LaTeX
plt.title(r'$\bf{Energy\ vs.\ Lattice\ Parameter}$ for Ti (Hexagonal, 2 atoms in basis)', fontsize=16)
plt.xlabel(r'Lattice parameter $a$ (in $\AA$)', fontsize=14)
plt.ylabel(r'Total Energy $E$ (in eV)', fontsize=14)

# Style tweaks
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# BCC data
e_bcc = np.array([
    0.88121809E+00, -0.90614961E+01, -0.13968050E+02, -0.14449979E+02, -0.14825838E+02,
    -0.15072803E+02, -0.15200717E+02, -0.15237351E+02, -0.15189393E+02, -0.15063952E+02,
    -0.14881779E+02, -0.14647649E+02, -0.14374294E+02, -0.12646800E+02, -0.10709479E+02, -0.88875022E+01
])
a_bcc = np.array([
    0.7, 0.8, 0.9, 0.92, 0.94, 0.96, 0.98,
    1.00, 1.02, 1.04, 1.06, 1.08,
    1.10, 1.20, 1.30, 1.40
])

# Hexagonal data
e_hex = np.array([
    -0.15512927E+02, -0.15380455E+02, -0.15123191E+02, -0.14722995E+02,
    -0.14158092E+02, -0.13419261E+02, -0.12490846E+02,
    -0.15540481E+02, -0.15480748E+02, -0.15384324E+02, -0.15253292E+02,
    -0.14611225E+02, -0.13759103E+02, -0.12798886E+02, -0.11798886E+02
])
a_hex = np.array([
    0.98, 0.96, 0.94, 0.92, 0.90, 0.88, 0.86,
    1.00, 1.02, 1.035, 1.05, 1.10, 1.15, 1.20, 1.25
])

# Sort both datasets
idx_bcc = np.argsort(a_bcc)
a_bcc_sorted = a_bcc[idx_bcc]
e_bcc_sorted = e_bcc[idx_bcc]

idx_hex = np.argsort(a_hex)
a_hex_sorted = a_hex[idx_hex]
e_hex_sorted = e_hex[idx_hex]

# Find minimums
i_bcc_min = np.argmin(e_bcc_sorted)
i_hex_min = np.argmin(e_hex_sorted)

a_bcc_min, e_bcc_min = a_bcc_sorted[i_bcc_min], e_bcc_sorted[i_bcc_min]
a_hex_min, e_hex_min = a_hex_sorted[i_hex_min], e_hex_sorted[i_hex_min]

# Plotting
plt.figure(figsize=(10, 6))

# BCC plot
plt.plot(a_bcc_sorted, e_bcc_sorted, color='royalblue', lw=2, label='BCC Titanium')
plt.scatter(a_bcc_sorted, e_bcc_sorted, s=50, color='skyblue', edgecolor='black')
plt.scatter(a_bcc_min, e_bcc_min, s=100, color='blue', edgecolor='black', zorder=5)


# Hexagonal plot
plt.plot(a_hex_sorted, e_hex_sorted, color='darkorange', lw=2, label='Hexagonal Titanium')
plt.scatter(a_hex_sorted, e_hex_sorted, s=50, color='orange', edgecolor='black')
plt.scatter(a_hex_min, e_hex_min, s=100, color='red', edgecolor='black', zorder=5)

# Annotate BCC minimum (move up-right)
# Annotate BCC minimum  Arrow pointing **down** (text above point)
plt.annotate(
    rf'$E_\mathrm{{BCC}} = {e_bcc_min:.2f}\ \mathrm{{eV}}$',
    xy=(a_bcc_min, e_bcc_min),
    xytext=(a_bcc_min, e_bcc_min + 1),  # text above, arrow down
    arrowprops=dict(arrowstyle='->', color='blue', lw=1.5),
    fontsize=12,
    color='black',
    ha='right'
)

# Annotate Hexagonal minimum  Arrow pointing **up** (text below point)
plt.annotate(
    rf'$E_\mathrm{{hex}} = {e_hex_min:.2f}\ \mathrm{{eV}}$',
    xy=(a_hex_min, e_hex_min),
    xytext=(a_hex_min, e_hex_min + 1),  # text below, arrow up
    arrowprops=dict(arrowstyle='->', color='red', lw=1.5),
    fontsize=12,
    color='black',
    ha='left'
)


plt.xlim([0.8,1.25])
plt.ylim(-16,-14)

# Axis labels and title
plt.title(r'$\bf{Comparison\ of\ BCC\ and\ Hexagonal\ Titanium}$', fontsize=16)
plt.xlabel(r'Lattice Parameter $a$ (in $\AA$)', fontsize=14)
plt.ylabel(r'Total Energy $E$ (eV)', fontsize=14)

# Final touches
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
